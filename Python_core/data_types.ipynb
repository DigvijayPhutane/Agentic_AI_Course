```python
#chapter 1 
#Not mutable example 
sugar_amount = 2
print(f"Initial sugar: {sugar_amount}")

sugar_amount = 22
print(f"second Initial sugar: {sugar_amount}")

print(f"ID of 2 : {id(2)}")
print(f"ID of 2 : {id(12)}")

# we have created a new reference|
```

```python
#chapter 2 
#Mutable 
spice_mix = set()
print(f"initial spice id : {id(spice_mix)}")
spice_mix.add("Ginger")
spice_mix.add("cardamom")
print(f"initial spice id : {id(spice_mix)}")
print(spice_mix)
#id didnt change
```

```python
#list is mutable

spice_mix_l = []
print(f"initial spice id : {id(spice_mix_l)}")
spice_mix_l.append("Ginger")
spice_mix_l.append("cardamom")
print(f"initial spice id : {id(spice_mix_l)}")
print(spice_mix_l)
```

```python
# Integers
black_tea_gram = 3
ginger_gram = 4
total_grams = black_tea_gram + ginger_gram
print(f"Total grams of tea ingredients: {total_grams}g")
```

```python
black_tea_gram = 10
ginger_gram = 4
total_grams = black_tea_gram // ginger_gram
print(f"Total grams of tea ingredients: {total_grams}g")
```

```python
black_tea_gram = 3
ginger_gram = 4
total_grams = black_tea_gram % ginger_gram
print(f"Total grams of tea ingredients: {total_grams}g")
# remainder 3 divided by 4 is 0 with remainder 3
```
